```python
# 검색하고자 하는 앱의 키워드 입력

search_word = input("검색하고자 하는 앱의 이름을 입력해주세요"). replace(',', '') # replace : 검색할떄 빈칸 있으면 삭제, 없어도 상관없음
url = 'https://play.google.com/store/search?q={}&c=apps&hl=ko'
url = url.format(search_word) 
#print(url)

browser.get(url)

time.sleep(3) # 로딩 후에 요소들 찾을 수 있기때문에, 로딩시간 줌
browser.implicitly_wait(5) # 로딩 후에 요소들 찾을 수 있기때문에, 로딩시간 줌

browser.find_element(By.CLASS_NAME, 'Si6A0c').click()
```

### 리뷰 가져오기

```python
# 1)리뷰 스크롤 3번 내리기 with time.sleep 2초 

browser.find_element(By.CLASS_NAME, 'fysCi').click() # 모달창 클릭(리뷰 전체화면, 스크롤 가능한곳)
modal = browser.find_element(By.CLASS_NAME, 'fysCi') # 내려줄 만큼의 높이를 = modal 로 정의
browser.execute_script('arguments[0].scrollTo(0,arguments[0]. scrollHeight);', modal)
                  

import time

for i in range(3):
    try:
        browser.execute_script("arguments[0].scrollTo(0, arguments[0].scrollHeight);", modal)  
        time.sleep(2)  
    except:
        print('try 에 오류가 났을 때, 여기로 옵니다')
        # browser.find_element(창 닫을 요소).click

## except : 예외처리(중간에 팝업창과 같은 상황 발행시, 예외처리 진행)                                      
                      
```

```python
import pandas as pd
```

```python
# 2) 데이터 가져오기 : 전체 데이터 가져온 후, 그 안에서 나눔

containers = browser.find_elements(By.CLASS_NAME, 'RHo1pe') # 전체 데이터

data_list = [] # 반복문 진행될떄마다 임시 저장! (for 문 안에서만 존재함) 

for container in containers :
    #print(container.text)
    
    # 유저 닉네임
    user_name = container.find_element(By.CLASS_NAME,'X5PpBb').text
    #print(user_name)
    
    # 평점
    rating = container.find_element(By.CLASS_NAME,'iXRFPc').get_attribute('aria-label')[10] 
    #aria-label="별표 5개 만점에 5개를 받았습니다."에서 10번쨰 문자 추출
    #print(rating)
    
    # 댓글 작성 날짜
    date= container.find_element(By.CLASS_NAME,'bp9Aid').text
    #print(date)
    
    # 댓글 내용
    content = container.find_element(By.CLASS_NAME,'h3YV2d').text
    
    #print('==='*10) 
    #print(content)

    # 텍스트-> 딕셔너리형태로 변환(제이슨 타입) : {키:밸류}
    ## DataFrame(변수)에서 변수는 리스트 형태만 가능
    data = {
        'user_name' : user_name,
        'rating' : rating,
        'date' : date,
        'content' : content
    }
    
    # print(data)
    data_list.append(data)
    
    
df = pd.DataFrame(data_list)
df.to_csv('./playstore_comment.csv', encoding='utf-8-sig') # csv 파일로 저장
df
```
